```python
import os
import numpy as np
import pandas as pd
from ultralytics import YOLO
```

#### Train on splits

```python
dataset_path = 'datasets/kfolds/yH2AX'
results_path = "results/kfold_yh2ax_cell"

model_name = 'yh2ax_cell'
results_file = 'results/results.csv'

n_splits = 5
batch = 16
epochs = 350
```

```python
ds_yamls = []
select_type = 'cell'

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        # Check based on the selected type
        if (select_type == 'foci' and file.endswith("foci.yaml")) or (select_type == 'cell' and file.endswith("cell.yaml")):
            ds_yamls.append(os.path.join(root, file))

ds_yamls
```

```python
results = {}

for k in range(n_splits):
    dataset_yaml = ds_yamls[k]
    model = YOLO('yolov8n.yaml')
    # model = YOLO('pretrained/YOLOv8m.pt')  # pretrained
    model.train(data=dataset_yaml, epochs=epochs, batch=batch, project=results_path, name=f'split_{k+1}')
    results[k] = model.metrics  # save output metrics for further analysis
```

---

## Evaluation

```python
cell = True

results_list = []

for k in range(n_splits):
    # Load the model for each fold
    model_path = f'{results_path}/split_{k+1}/weights/best.pt'
    model = YOLO(model_path)
    
    # Evaluate on the validation dataset
    results = model.val()
    
    # Get the confusion matrix
    conf_matrix = results.confusion_matrix
    conf_matrix_np = conf_matrix.matrix.astype(np.int32)
    
    # Calculate metrics for this fold
    TP = conf_matrix_np.diagonal()
    FN = conf_matrix_np.sum(axis=0) - TP
    FP = conf_matrix_np.sum(axis=1) - TP
    TN = conf_matrix_np.sum() - (FP + FN + TP)
    
    precision = TP / (TP + FP + 1e-6)  # Avoid division by zero
    recall = TP / (TP + FN + 1e-6)
    accuracy = (TP + TN) / (TP + FP + FN + TN + 1e-6)
    
    if cell:
        results_list.append({
        'model': f'{model_name}{k+1}',
        'tp': TP[0],
        'fp': FP[0],
        'fn': FN[0],
        'precision': precision[0],
        'recall': recall[0],
        'accuracy': accuracy[0],
        'tp_cell': TP[1],
        'fp_cell': FP[1],
        'fn_cell': FN[1],
        'precision_cell': precision[1],
        'recall_cell': recall[1],
        'accuracy_cell': accuracy[1]
    })
    else:
        results_list.append({
            'model': f'{model_name}{k+1}',
            'tp': TP[0],
            'fp': FP[0],
            'fn': FN[0],
            'precision': precision[0],
            'recall': recall[0],
            'accuracy': accuracy[0]
        })

df = pd.DataFrame(results_list)

if cell:
    results_file = 'results/results_cell.csv'
df1 = pd.read_csv(results_file)

merged_df = pd.concat([df1, df], ignore_index=True)
merged_df = merged_df.sort_values(by='model', ascending=True)

merged_df.to_csv(results_file, index=False)

# Average metrics across folds
avg_metrics = {
    'tp': np.mean([res['tp'] for res in results_list], axis=0),
    'fp': np.mean([res['fp'] for res in results_list], axis=0),
    'fn': np.mean([res['fn'] for res in results_list], axis=0),
    'precision': np.mean([res['precision'] for res in results_list], axis=0),
    'recall': np.mean([res['recall'] for res in results_list], axis=0),
    'accuracy': np.mean([res['accuracy'] for res in results_list], axis=0)
}

# Print average metrics
print("Average Metrics Across Folds:")
print(f"True Positives: {avg_metrics['tp']}")
print(f"False Positives: {avg_metrics['fp']}")
print(f"False Negatives: {avg_metrics['fn']}")
print(f"Precision: {avg_metrics['precision']}")
print(f"Recall: {avg_metrics['recall']}")
print(f"Accuracy: {avg_metrics['accuracy']}")
```
